##UNIVARIATE ANALYSIS BIVARIATE AND MULTIVARIATE

With Univariate Analysis, only One variable is Involved

While Bivariate Analysis gives the relationship between the two variables

And Multivariate Analysis gives/shows the Relationship for so many variables

Some questions to consider when getting started can include:
1. How many variables do I have?

2. Do I have missing data?

3. What types of variables do I have?

There are two types of Variables:
1. Categorical Data

This type of data can be further broken down into nominal, ordinal, and binary values.
Ordinal Has Order say from Small to Big or Low to High
Nominal has no order
Binary has true or false or 1/0

2. Numerical Data

This type of Data is divided into two:
Discrete: These variables contain Finite/Whole Numbers
Continous: These variables contain Infinite Numbers

## BIVARIATE 
This gives the relation ship between the two variables , hence its called bivariate analysis

```python
##Box Plot
sns.boxplot(x='Surv_status', y='Age', data=Cancer)
plt.show()
```

```python
##Operation Year
sns.boxplot(x='Surv_status', y='Operation_year', data=Cancer)
plt.show()
```

```python
###Voilin Plots
##Voilin plots also like box plots, but these give pdf along with box plots in it. they look a voilin
## you can observe that people with operation year from 58 to 66 survived 
sns.violinplot(x='Surv_status', y='Operation_year', data=Cancer)
plt.show()
```

```python
plt.scatter(x='Surv_status', y='Age', data=Cancer)
plt.show()
```

```python
plt.scatter(x='Surv_status', y='Operation_year', data=Cancer)
plt.show()
```

##MULTIVARIATE ANALYSIS
Pair plot shows a clear and nice view of all variables and their realtion ship with all other variables

```python
##Pair Plot
sns.pairplot(Cancer, hue="Surv_status", diag_kind='hist')
##N.B If I don't specify the diag_kind argument, it will try to draw a kde and create an error.
```
